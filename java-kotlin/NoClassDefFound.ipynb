# Problem:

---




Recently, I have faced this problem. I was learning Kotling programming. I am using IntelliJ for this purpose.
While reading about <a href="https://kotlinlang.org/docs/java-interop.html">**Interoperability**</a>, I came across the problem.

```python
Exception in thread "main" java.lang.NoClassDefFoundError: MyJava
	at MyKotlinKt.main(MyKotlin.kt:5)
Caused by: java.lang.ClassNotFoundException: MyJava
	at java.net.URLClassLoader.findClass(URLClassLoader.java:382)
	at java.lang.ClassLoader.loadClass(ClassLoader.java:418)
	at sun.misc.Launcher$AppClassLoader.loadClass(Launcher.java:355)
	at java.lang.ClassLoader.loadClass(ClassLoader.java:351)
	... 1 more

Process finished with exit code 1
```

Here is my Java code.

```python
public class MyJava {

	public static void main(String[] args){

	}

	public static int getArea(int length,int breadth){

		return length*breadth;
	}

	public int add(int a, int b){

		return (a+b);
	}
}
```

Here is the Kotlin code

```python
fun main(args: Array<String>){

    val obj= MyJava()
    var area:Int= MyJava.getArea(10,5)
    var sum:Int= obj.add(10,30)
    println("Call static method => Area is : $area")
    println("Call instance method => Sum is : $sum")
}
```

It may seem to you that there is a problem with my build.gradle.kts file.
Please go through this <a href="https://discuss.kotlinlang.org/t/noclassdeffound-with-kotlin-java-project/20458">link</a> also. My problem is similar to this.

# Solution:

---



**Step 1:** Open the project in your IntelliJ IDEA ide.
Focus on the red markable site.

Screenshot (217).png

Here we will see we are inside the **kotlin** directory and here we have a package named **kotlinTutorial** and inside this package, we have our two files.

**Step 2:** Create a new directory named **`java`**

Right click on main folder => New => Directory => set directory name as "java" => Enter


Screenshot (219).png

**Step 3:** Cut your **java file** `(MyJava.java)` from the kotlinTutorial package and paste it inside the **java** directory.

**Step 4:** Run your application.
